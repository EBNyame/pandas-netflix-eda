## **Handle Missing Values**

```python
# checking if there is NaN in the dataset
df.isnull()
```

```python
# checking how many missing values are in each column
df.isnull().sum()
```

```python
# since I dont know some of the directors of the movies, I will set all the unkown directors to 'Not Available' because some netflix contents does not list a director
df["director"].fillna("Not Available", inplace=True)
```

```python
df
```

```python
# changing the NaN values to 'Unkown' in country's column
df["country"].fillna("Unkown", inplace=True)
```

```python
df
```

```python
# made a mistake; changing 'Unkown' to 'Not Available' in director's column
df["director"] = df["director"].replace("Unkown","Not Available")
```

```python
df
```

```python
df["date_added"].info()
```

```python
# converting the 'date_added' column from plain text to real date objects
df["date_added"] = pd.to_datetime(df['date_added'], format='mixed')
```

```python
df["date_added"].dtype
```

```python
df["date_added"].dt.year
```

```python
df
```

```python
# extracting the year from each date and also remove the decimal and the number after the decimal
df["year_added"] = pd.to_datetime(df["date_added"]).dt.year.astype("Int64")
```

```python
df
```

```python
# Total movies and total TV Shows are on Netflix
df["type"].value_counts()
```

```python
# Countries that produce the most Netflix content
df["country"].value_counts().head(5)
```

```python
# How Netflix contents grow over time
df["year_added"].value_counts().sort_index()
```

```python
# Most common age rating on Netflix
df["rating"].value_counts()
```

```python
# Save cleaned Data
df.to_csv("cleaned_netflix_data.csv", index=False)
```

## **Visualization**

**Movies Vs TV Shows**

```python
import matplotlib.pyplot as plt
```

```python
df['type'].value_counts().plot(kind= 'bar')
plt.title("Movies vs TV Shows")
plt.xlabel('Type')
plt.ylabel('Count')
plt.show();
```

**Content added per year**

```python
year_counts = df["year_added"].value_counts().sort_index()
year_counts.plot(kind= 'line')
plt.title("Content Added Over the Years")
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show();
```

**Ratings Distribution**

```python
df['rating'].value_counts().sort_index().plot(kind='bar')
plt.title("Content Ratings Distribution")
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
