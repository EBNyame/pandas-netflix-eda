## **Handle Missing Values**

```python
# checking if there is NaN in the dataset
df.isnull()
```

```python
# checking how many missing values are in each column
df.isnull().sum()
```

```python
# since I dont know some of the directors of the movies, I will set all the unkown directors to 'Not Available' because some netflix contents does not list a director
df["director"].fillna("Not Available", inplace=True)
```

```python
df
```

```python
# changing the NaN values to 'Unkown' in country's column
df["country"].fillna("Unkown", inplace=True)
```

```python
df
```

```python
# made a mistake; changing 'Unkown' to 'Not Available' in director's column
df["director"] = df["director"].replace("Unkown","Not Available")
```

```python
df
```

```python
df["date_added"].info()
```

```python
# converting the 'date_added' column from plain text to real date objects
df["date_added"] = pd.to_datetime(df['date_added'])
```

```python
df["date_added"].dtype
```

```python
df["date_added"].dt.year
```

```python
df
```

```python
# extracting the year from each date and also remove the decimal and the number after the decimal
df["year_added"] = pd.to_datetime(df["date_added"]).dt.year.astype("Int64")
```
